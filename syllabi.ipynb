```python
#data read
drs = ["prof1", "prof2", "prof3", "prof4", "prof5"]
#["denlinger", "england", "spicer", "walters", "wheeler"]
syllabi = []

def txtread(filename, encode="utf-8", lower=True):
    """after reading a file (expected to be .txt) saved as 'filename' & encoded as 'encode',
    return the result as a one-line string.
    """
    with open(filename, encoding=encode) as f:
        text = f.read()
        if lower:
            text = text.lower()
    return text
    
for prof in drs:
    filename = prof + ".txt"
    syllabus = txtread(filename)
    syllabi.append(syllabus)
        
print(len(syllabi))
for i in range(5):
    print(len(syllabi[i]))
```

```python
#tokenize
import nltk
from nltk.corpus import stopwords

stopwords = stopwords.words('english')
"from" in stopwords
```

```python
import nltk

def tokenize(text, stopwords=True):
    """
    after reading 'text' thru the txtread() function,
    return the result as a list of tokens.
    """
    text_tokenized = nltk.word_tokenize(text)
    
    if (stopwords == "False"):
        return text_tokenized    

    else:
        stopwords = nltk.corpus.stopwords.words('english')
        tokens = []
        for token in text_tokenized:
            if (token not in stopwords):
                tokens.append(token)
        return tokens

cnt = 0
syl_tokenized = []

for syllabus in syllabi:
    syl_tokenized.append(tokenize(syllabus))   
    print(f"tokenization complete: {drs[cnt]}")
    cnt += 1
```

```python
#func check: tokenize()
for i in range(5):
    print(syl_tokenized[i][-30:])
```

```python
def freqdict(tokens, sort=True):
    """
    after receiving a list of tokens,
    return a dictionary having type of tokens as keys and freq of tokens as value.
    """
    freqdict = {}
    for token in tokens:
        if (token not in freqdict.keys()) :
            freqdict[token] = 1
        else:
            freqdict[token] += 1
            
    if (sort):
        for voca, freq in freqdict.items():
            freqdict_sorted = sorted(freqdict.items(), reverse = True, key = lambda x: x[1])  
        return freqdict_sorted
    
    return freqdict
```

```python
ut = txtread("wiki_utexas.txt")
ut_tokenized = tokenize(ut)
ut_freqdict = freqdict(ut_tokenized, sort=False)
ut_freqdict
```

```python
#us_freq: {prof1 :  {w1: f1, w2: f2}, prof2 : {w1:f1, w2:f2}}
us_freq = {}
cnt = 0

for syl in syl_tokenized:
    wordfreq = freqdict(syl)
    us_freq[f"{drs[cnt]}"] = wordfreq
    print(f"frequency counting complete: {drs[cnt]}")
    cnt += 1
```

```python
topnum = 50

print(f"=======TOP {topnum} FREQUENT WORDS in LIN 306 Syllabi=======")
print()

print("=======Sorted by: Raw Freq=======")

total = ""
for syllabus in syllabi:
    total += syllabus
syl_tokenized_total = tokenize(total)
freqdict_total = freqdict(syl_tokenized_total)
print(freqdict_total[:topnum])
print()

print("=======Sorted by: Prof=======")
for dr in drs:
    print(dr)
    print(us_freq[dr][:topnum])
    print()
```
